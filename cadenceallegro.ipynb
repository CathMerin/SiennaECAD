```python
import pandas as pd
import xlrd
final_data = list()
column_one = ""
column_two = ""
# fileName1 = input("Enter the name of the file to convert\n")
wb = xlrd.open_workbook("CadenceAllegroTest.xls")
sheet = wb.sheet_by_index(0)
for i in range(3, sheet.nrows):
    column_one = sheet.cell_value(i, 0)
    cmps = sheet.cell_value(i, 1).split()
    for j in cmps:
        column_two = j
        temp_dict = {
            "column_one": column_one,
            "column_two": column_two
        }
        final_data.append(temp_dict)
print(final_data)
```
